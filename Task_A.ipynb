**A: Identify and compare the top 10 groups of visitors at each theme park branch according to their location of origin.**

```python
# Taking the (columns: Branch and Reviewer_Location) from the df and grouping them up to count the no. of reviews
visitor_groups = df.groupby(['Branch', 'Reviewer_Location']).size().reset_index(name='Count')

# Display the result to understand the structure
visitor_groups.head()
```

```python
#dimensions
print(visitor_groups.shape)
```

```python
# for getting the top ten groups for each branch (California, HongKong, Paris)
top_10_visitors = visitor_groups.groupby('Branch').apply(lambda x: x.nlargest(10, 'Count')).reset_index(drop=True)

# Top 10 visitor groups for each branch
top_10_visitors
```

```python
# importing the libraries for visualisations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visual to the top 10 visitor groups for each branch (Graph)
plt.figure(figsize=(14, 8))
sns.barplot(data=top_10_visitors, x='Count', y='Reviewer_Location', hue='Branch', dodge=True)
plt.title('Fig 1: Top 10 Visitor Groups by Branch')
plt.xlabel('Number of Reviews')
plt.ylabel('Visitor Location')
plt.legend(title='Branch')
plt.show()
```

A. The figure 1 shows the top 10 visitor groups by branch. Analysing the figure the following are the key takeaways in terms of differences in the top visitor groups among the branches:
- The branch with the most visitors is the California.
- California branch has most of the visitors from the United States numbering around 10000.
- The maximum number of visitors for the Paris branch i.e. around 6500, are from United Kingdom
- HongKong branch has most of its visitors from Australia counting to more than 2000
- Its a trend for the visitor group to visit the nearest branch for instance the groups from the US visited the California branch, those from UK visited the Paris branch and those from Australia visited the HongKong branch.
- Its astonishing to see that groups from:
  - UK visited more than those from France to the Paris branch and
  - Australia visited more than those from HongKong to the HongKong branch.
