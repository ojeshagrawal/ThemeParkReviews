**A: Identify and compare the top 10 groups of visitors at each theme park branch according to their location of origin.**

```python
# Taking the (columns: Branch and Reviewer_Location) from the df and grouping them up to count the no. of reviews
visitor_groups = df.groupby(['Branch', 'Reviewer_Location']).size().reset_index(name='Count')

# Display the result to understand the structure
visitor_groups.head()
```

```python
#dimensions
print(visitor_groups.shape)
```

```python
# for getting the top ten groups for each branch (California, HongKong, Paris)
top_10_visitors = visitor_groups.groupby('Branch').apply(lambda x: x.nlargest(10, 'Count')).reset_index(drop=True)

# Top 10 visitor groups for each branch
top_10_visitors
```
